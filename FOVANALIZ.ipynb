```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def scale_image(img1, scale_factor):
    # Масштабируем изображение
    height, width = img1.shape[:2]
    new_dim = (int(width * scale_factor), int(height * scale_factor))
    scaled_img = cv2.resize(img1, new_dim, interpolation=cv2.INTER_LINEAR)
    return scaled_img

# Загрузка изображений
image1_path = '/content/photo_2024-12-05_11-29-53 (1).jpg'  # Путь к первому изображению
image2_path = '/content/img31.png'  # Путь ко второму изображению

img1 = cv2.imread(image1_path)
img2 = cv2.imread(image2_path)

# Расстояние между камерами (в см)
distance_between_cameras = 10

focal_length_camera1 = 20
focal_length_camera2 = 1

# Вычисление коэффициента масштабирования
scale_factor = focal_length_camera1 / focal_length_camera2

# Масштабируем первое изображение
scaled_img1 = scale_image(img1, scale_factor)

# Отображаем результаты
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
plt.title("Первое изображение (исходное)")
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Первое изображение, масштабированное для совпадения")
plt.imshow(cv2.cvtColor(scaled_img1, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
